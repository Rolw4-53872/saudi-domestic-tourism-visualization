```python
import pandas as pd

# Load the CSV file into a pandas DataFrame, specifying the semicolon as a separator
df = pd.read_csv('/content/بيانات عن السياحة.csv', sep=';')

# Display the first 5 rows of the DataFrame
display(df.head(30850))
```

```python
# Display information about the DataFrame, including data types and non-null counts
df.info()
```

```python
import pandas as pd

df = df.copy()

# 1) mean per (province + month)
dest_month_mean = df.groupby(['destinationProvinceNameEn', 'month'])['destination_temp'].transform('mean')
orig_month_mean = df.groupby(['originProvinceNameEn', 'month'])['origin_temp'].transform('mean')

# 2) mean per province only
dest_province_mean = df.groupby('destinationProvinceNameEn')['destination_temp'].transform('mean')
orig_province_mean = df.groupby('originProvinceNameEn')['origin_temp'].transform('mean')

# 3) overall mean
dest_global_mean = df['destination_temp'].mean()
orig_global_mean = df['origin_temp'].mean()

# fill destination_temp
df['destination_temp'] = (
    df['destination_temp']
    .fillna(dest_month_mean)
    .fillna(dest_province_mean)
    .fillna(dest_global_mean)
)

# fill origin_temp
df['origin_temp'] = (
    df['origin_temp']
    .fillna(orig_month_mean)
    .fillna(orig_province_mean)
    .fillna(orig_global_mean)
)

df.info()
```

```python
# 3) تصدير الملف بعد التنظيف
df.to_csv("/content/داتاسيت خاصة بمشروع مادة تصوير البيانات النهائي.csv", index=False)

print("تم التنظيف وتصدير الملف بنجاح 👍")
```
